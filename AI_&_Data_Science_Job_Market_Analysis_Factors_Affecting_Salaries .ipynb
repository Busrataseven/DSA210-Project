##Hypothesis Testing: Experience Level and Salary

To statistically test whether experience level has a significant effect on salary, a one-way ANOVA test is conducted.

H0: There is no difference in salary across experience levels.  
H1: At least one group has a different mean salary.

The high p-value indicates that there is no statistically significant difference between remote and onsite salaries. This reinforces the idea that work type does not independently influence salary levels.

```python
from scipy.stats import f_oneway

entry_salary = df[df["experience_level"] == "Entry"]["salary"]
mid_salary = df[df["experience_level"] == "Mid"]["salary"]
senior_salary = df[df["experience_level"] == "Senior"]["salary"]

f_stat, p_value = f_oneway(entry_salary, mid_salary, senior_salary)

print("F-statistic:", f_stat)
print("p-value:", p_value)
```

The ANOVA test shows that there is a statistically significant difference in salary across experience levels (p < 0.05).

While experience level shows a strong and statistically significant effect on salary, it is also important to examine whether other factors, such as work type (remote, hybrid, onsite), contribute to salary differences.

Salary vs Work Type

This section examines how salary differs across remote, hybrid, and onsite jobs.

This suggests that remote work alone is not a strong determinant of salary and that companies may offer similar compensation regardless of work arrangement.

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x="remote_type",
    y="salary",
    hue="remote_type",
    data=df,
    palette="Set2",
    legend=False
)

plt.title("Salary vs Work Type", fontsize=14)

plt.show()
```

##Average Salary by Work Type

This chart shows the average salary for each work type to better compare overall differences.

The bar chart shows that the average salaries across remote, hybrid, and onsite roles are quite similar. This indicates that work type alone may not be a strong determinant of salary. Therefore, other factors such as experience level or skills may play a more significant role in explaining salary differences.

```python
df.groupby("remote_type")["salary"].mean().plot(
    kind="bar",
    color=["#4CAF50", "#2196F3", "#FF9800"]
)

plt.title("Average Salary by Work Type")
plt.ylabel("Average Salary")
plt.xticks(rotation=0)
plt.show()
```

This observation is consistent with the hypothesis test results, which showed no statistically significant difference between remote and onsite salaries.

The t-test results show that there is no statistically significant difference between remote and onsite salaries (p ≥ 0.05). Therefore, we fail to reject the null hypothesis. This suggests that work type does not have a significant effect on salary.

```python
from scipy.stats import ttest_ind

remote = df[df["remote_type"] == "Remote"]["salary"]
onsite = df[df["remote_type"] == "Onsite"]["salary"]

t_stat, p_value = ttest_ind(remote, onsite)

print("t:", t_stat)
print("p:", p_value)
```

## Data Enrichment with GDP per Capita

To enrich the analysis, GDP per capita data from the World Bank is added to the original dataset. This helps examine whether country-level economic conditions are related to salary levels.

```python
gdp_df = pd.read_csv("API_NY.GDP.PCAP.CD_DS2_en_csv_v2_121663.csv", skiprows=4)

gdp_df = gdp_df[["Country Name", "2023"]].rename(
    columns={"Country Name": "country", "2023": "gdp_per_capita"}
)

gdp_df = gdp_df.dropna(subset=["gdp_per_capita"])

gdp_df.head()
```

### Merging GDP Data

The GDP dataset is merged with the main job market dataset using the country column.

```python
df_enriched = df.merge(gdp_df, on="country", how="left")

print("Missing GDP values:", df_enriched["gdp_per_capita"].isnull().sum())
```

```python
df_enriched["gdp_per_capita"] = df_enriched["gdp_per_capita"].fillna(
    df_enriched["gdp_per_capita"].median()
)

print("Missing GDP values after filling:", df_enriched["gdp_per_capita"].isnull().sum())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df_enriched["gdp_per_capita"], bins=30, kde=True)

plt.title("Distribution of GDP per Capita")
plt.xlabel("GDP per Capita")
plt.ylabel("Frequency")
plt.show()
```

The GDP per capita distribution appears highly skewed, with most values concentrated in a mid-range and a smaller number of countries having very high GDP levels.

This indicates that the dataset includes countries with very different economic conditions. However, since the distribution is not uniform, this may affect how GDP relates to salary in the analysis.

```python
plt.figure(figsize=(10, 6))
df_enriched["country"].value_counts().plot(kind="bar")

plt.title("Country Distribution in the Dataset")
plt.xlabel("Country")
plt.ylabel("Number of Job Postings")
plt.xticks(rotation=45)
plt.show()
```

The country distribution shows that the dataset is not perfectly balanced across countries. Some countries appear more frequently than others.

This imbalance may influence the analysis, as countries with more observations will have a stronger effect on overall patterns. Therefore, results related to country-level variables such as GDP per capita should be interpreted with caution.

The dataset is not evenly distributed across countries. Some countries appear more frequently than others, which may influence the overall relationship between GDP per capita and salary.

### GDP per Capita and Salary

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_enriched, x="gdp_per_capita", y="salary")

plt.title("GDP per Capita vs Salary")
plt.xlabel("GDP per Capita")
plt.ylabel("Salary")
plt.show()
```

```python
df_enriched[["salary", "gdp_per_capita"]].corr()
```

The relationship between GDP per capita and salary appears weak, with a near-zero correlation value. This suggests that GDP per capita alone does not strongly explain salary differences in this dataset.

The GDP distribution and country occurrence plots also show that the dataset is imbalanced across countries. This imbalance may limit the ability to detect clear patterns between GDP and salary.

Therefore, while GDP per capita is an important macroeconomic indicator, it does not appear to be a strong predictor of salary in this dataset.

```python
numeric_corr = df_enriched.corr(numeric_only=True)

plt.figure(figsize=(10, 6))
sns.heatmap(numeric_corr, annot=True, cmap="coolwarm", fmt=".2f")

plt.title("Correlation Matrix of Numerical Variables")
plt.show()
```

The correlation matrix provides several important insights:

- Years of experience shows a positive relationship with salary, indicating that more experienced individuals tend to earn higher salaries.
- Skills such as machine learning and cloud technologies have moderate positive correlations with salary, suggesting that these skills may contribute to higher-paying roles.
- Basic skills like SQL and Python show weaker correlations, possibly because they are common across many roles regardless of salary level.
- GDP per capita has almost no correlation with salary, reinforcing the idea that country-level economic indicators do not strongly explain salary differences in this dataset.

Overall, these results suggest that experience and specialized technical skills are more important drivers of salary than general skills or macroeconomic factors.
